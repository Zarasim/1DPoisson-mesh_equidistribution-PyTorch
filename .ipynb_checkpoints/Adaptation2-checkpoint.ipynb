Choose parameters for miniminising interpolation error

k: degree of interpolating piecewise polynommials

m: Order of derivatives in the Hilbert space (Sobolev space with p=2)  0 <= m <= k

For k = 1 and m = 0

Error bound is Ciarlet  |u - u_k|^2_H^m <= C sum h_j^5 |u|^2_H^2  if u in H^2

For simplicity replace |u| by evaluating u directly at the midpoint of the interval h_j.

The output of the network must be the transformation map xi -> x

The loss function must be computed passing all the coordinates for each epoch without mini_batches.

## Define poissible classes for training the network

```python
class softplus_power(torch.nn.Module):
    def __init(self):
        super.__init__()

    def forward(self, x):
        m = nn.Softplus()
        x = m(x)
        return x**1.1
```

```python
class ICNN(nn.Module):
    def __init__(self, n_inputs, n_outputs, n_layers, n_npl,lb,ub):
        super().__init__()
        self.n_inps = n_inputs
        self.lb = lb
        self.ub = ub
        if n_layers == 0:
            self.layers = nn.ModuleList()
            self.activations = nn.ModuleList()
            self.layers.append(nn.Linear(n_inputs, n_outputs))
        else:
            self.layers = nn.ModuleList()
            self.activations = nn.ModuleList()

            self.layers.append(nn.Linear(n_inputs, n_npl))
            self.activations.append(nn.Tanh())

            for x in range(n_layers-1):
                # add n_inputs as shown in Fig.1 in the paper
                self.layers.append(nn.Linear(n_npl + n_inputs, n_npl))
                self.activations.append(nn.Tanh())
            self.layers.append(nn.Linear(n_npl + n_inputs, n_outputs))

    '''
    Forward pass - Simple fully connected network with given activations
    '''

    def forward(self, x):
        inps = x
        out = x
        for i in range(len(self.layers)-1):
            if i > 0:
                # stack out and inps along dim = 1 as hstack
                # from the second layer the activaction function is activated on nÂ° nodes: inps + out
                out = torch.cat((out, inps), dim=1)
            out = self.layers[i](out)
            out = self.activations[i](out)

        if len(self.layers) != 1:
            out = torch.cat((out, inps), dim=1)

        # out is of shape (ndata,1)
        out = self.layers[-1](out)
        out = rescale(out,self.lb,self.ub)    
        return out
        
    def make_convex(self):
        with torch.no_grad():
            for i in range(1, len(self.layers)):
                if i == 0:
                    self.layers[i].weight[:, :][self.layers[i].weight[:, :] < 0] = \
                        torch.abs(self.layers[i].weight[:, :]
                                  [self.layers[i].weight[:, :] < 0])
                self.layers[i].weight[:, :-self.n_inps][self.layers[i].weight[:, :-self.n_inps] < 0] = \
                    torch.abs(self.layers[i].weight[:, :-self.n_inps]
                              [self.layers[i].weight[:, :-self.n_inps] < 0])

    '''
    Sets all negative weights (W^z) to 0
    '''

    def project(self):
        with torch.no_grad():
            for i in range(1, len(self.layers)):
                if i == 0:
                    self.layers[i].weight[:,
                                          :][self.layers[i].weight[:, :] < 0] = 0
                self.layers[i].weight[:, :-
                                      self.n_inps][self.layers[i].weight[:, :-self.n_inps] < 0] = 0        
    
    
    def save(self, path, model_name):
        # Python dictionary object that maps each layer to its parameter tensor.
        torch.save(self.state_dict(), path + model_name + '.pth')

    '''
    Loads the model given its path.
    Different Architecture or activation function (with parameters) in the saved model will throw an error

    Load:
    model = TheModelClass(*args, **kwargs)
    model.load_state_dict(torch.load(PATH))
    model.eval()
    '''

    def load(self, path):
        # load the parameters from pth file
        self.load_state_dict(torch.load(path))
```

```python
# the deep neural network
class DNN(torch.nn.Module):
    
    # layers count the hiddens + the output 
    def __init__(self, layers, lb, ub):
        
        super().__init__()
        
        # parameters
        self.depth = len(layers) - 1
        self.lb = lb
        self.ub = ub
        #self.activation = softplus_power
        #self.activation = nn.Tanh
        
        # set up layer order dict
        self.layers = torch.nn.ModuleList()
        self.activations = torch.nn.ModuleList()
        
        for i in range(self.depth - 1): 
            self.layers.append(torch.nn.Linear(layers[i], layers[i+1]))
            self.activations.append(nn.Tanh())
            
        self.layers.append(nn.Linear(layers[-2], layers[-1]))
    
                                     
    def forward(self, x):
        
        out = x
        for i in range(len(self.layers)-1):
            out = self.layers[i](out)
            out = self.activations[i](out)

        # out is of shape (ndata,1)
        out = self.layers[-1](out)
        out = rescale(out,self.lb,self.ub)    
        return out
```

```python
def rescale(x,lb,ub):
    """ rescaling function to impose boundary conditions """           
    torch.sort(x)[0]
    y = x - torch.min(x) 
    z = y.squeeze()/(torch.max(y).squeeze())
    z = lb + (ub - lb)*z
    
    return z 
```

```python
class PhysicsInformedNN():

    def __init__(self, xi, u, layers , k , m):
        
        # data
        self.xi = torch.tensor(xi, requires_grad=True).float().to(device).unsqueeze(-1)
        self.u = torch.tensor(u).to(device) # defualt requires_grad: false
        self.N = len(self.xi)
        # boundary conditions
        self.lb = self.xi[0].clone().detach()
        self.ub = self.xi[-1].clone().detach().requires_grad_(True)
        
        # deep neural networks
        self.layers = layers
        #self.dnn = DNN(self.layers,self.lb,self.ub).to(device)
        self.dnn = ICNN(*layers,self.lb,self.ub).to(device)
        
        # optimizers: using the same settings
        #self.optimizer = torch.optim.Adam(self.dnn.parameters(),lr = 0.1)
        self.optimizer = torch.optim.Adadelta(self.dnn.parameters(),lr = 0.1)
        #self.optimizer = torch.optim.LBFGS(
        #    self.dnn.parameters(),   # derived from class nn.Module
            #lr = 1,                  # default is 1
            #max_iter=10, 
            #max_eval=10, 
            #history_size=50,
            #tolerance_grad=1e-10, 
            #tolerance_change=1.0 * np.finfo(float).eps,   # Machine limits for floating point types.
            #line_search_fn="strong_wolfe"       # can be "strong_wolfe" worse result
        #)
        
        self.k = k
        self.m = m

        self.iter = 0
    
    def second_derivative(self,xi,x,u):
        
        """ Compute second derivative given res """
        
        # convert p to numpy array
        x_np = x.detach().cpu().numpy()
        xi_np = x.detach().cpu().numpy()
        u_np = u.detach().cpu().numpy()
        
        # interpolate with scipy and compute second derivative at the required points
        f = Interp(xi_np,u_np)
        f_x = f.derivative()
        f_xx = f_x.derivative()
        # evaluate second derivative at thr middle points of the equidistributed mesh points x 
        x_np = (x_np[1:] + x_np[:-1])/2
        integral = f_x(x_np) 
        
        #f = Interp(xi_np,u_np)
        #f = f(x_np)
        #f_xx = f[:-1].copy()
        #eps = 1e-5
        #hi = (x_np[1:] - x_np[:-1]) + eps 
        
        #f_xx[0] = (f[1] - f[0])/hi[0]
        #for i in range(1,len(f_xx)-1):
            #f_xx[i] = (f[i+1] - 2*f[i] + f[i-1])/(hi[i]**2) 
         #   f_xx[i] = (f[i+1] - f[i])/hi[i] 
        #f_xx[-1] = (f[-2] - f[-1])/hi[-1]
        #integral = f_xx
        
        return integral
        
    def net_interp_bound(self,xi,u):
        
        """ Compute Interpolation upper bound """
        
        x = self.dnn(xi)
        #print(x)
        h = x.squeeze()
        h = h[1:] - h[:-1]
        
        integral = self.second_derivative(xi,x,u)
        integral = torch.tensor(integral).float().to(device)
        integral = torch.pow(integral,2).squeeze()
        
        alpha = 1.0
        #f = h*(alpha*integral)**(1/(1+2*(self.k - self.m+1)))
        #f = torch.pow(f,1+2*(self.k-self.m+1))
        
        f = h**(1+2*(self.k-self.m+1))*integral
        
        return f,x
    
    def loss_func(self):
        alpha = 1
        self.optimizer.zero_grad()
        f_pred,x = self.net_interp_bound(self.xi,self.u)
        #loss_f = ((self.N-1)**(2*(self.k-self.m+1)))*alpha*torch.sum(f_pred)
        loss_f = torch.sum(f_pred)
        #loss_u = torch.sum(torch.pow(self.u[0],2) + torch.pow(self.u[-1],2))
        #loss_xi = torch.sum(torch.pow(self.lb - x[0],2) + torch.pow(self.ub - x[-1],2))
        
        loss = loss_f
        #+ loss_xi + loss_u
        loss.backward()
        print('Iter %d, Loss: %.5e' % (self.iter, loss.item()))
        return loss
    
    def train(self):
        self.dnn.train()
        self.iter = self.iter + 1   
        # Backward and optimize
        self.optimizer.step(self.loss_func)

            
    def predict(self, Xi):
        
        Xi = torch.tensor(Xi, requires_grad=True).float().unsqueeze(-1).to(device)
        self.dnn.eval()
        x = self.dnn(Xi)
        x = rescale(x,self.lb,self.ub)
        return x
```

```python
# create uniform mesh with N points
N = 100
a = 0
b = 2*np.pi

xi = np.linspace(a,b,N)
u = np.sin(xi)
```

### Training

```python
layers_DNN = [1, 128, 256, 512, 256, 128, 1]
layers_ICNN = [1, 1, 5, 128]
model_ICNN = 1

if model_ICNN:
    net = ICNN(*layers_ICNN,a,b)
    dic = net.state_dict()
    for k in dic:
        dic[k] *= 0
    net.load_state_dict(dic)
    del(dic)
else:
    net = DNN(layers_DNN,a,b)
    dic = net.state_dict()
    for k in dic:
        dic[k] *= 0
    net.load_state_dict(dic)
    del(dic)

k = 1     # order of interpolation 
m = 0     
p = 2     # For l2 error


model = PhysicsInformedNN(xi, u, layers_ICNN, k ,m)
```

```python
%%time
 
net.make_convex()

# define n_epochs and train for that epochs 
nepochs = 5000
for _ in range(nepochs):
    model.train()
    net.project()
```
